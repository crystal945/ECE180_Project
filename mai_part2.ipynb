```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/' + 'Price2Rent_Ratio' + '.csv')
```

```python
df
```

```python
df.iloc[3,7:][::12]
```

```python
plt.figure(figsize=(17,10))
plt.plot(df.iloc[3,7:][::12])
plt.title('Price to Rent Ratio, San Diego County', fontsize = 20)
plt.xlabel('Year', fontsize = 16)
plt.ylabel('Ratio [%]', fontsize = 16)
plt.show()
```

```python
plt.figure(figsize=(17,10))
sns.countplot(df.iloc[3,7:][::12])
plt.title('Price to Rent Ratio, San Diego County', fontsize = 20)
plt.xlabel('Year', fontsize = 16)
plt.ylabel('Ratio [%]', fontsize = 16)
plt.show()
```
