```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.dates
from pandas import Series, DataFrame, Panel
import seaborn as sns
```

## https://fred.stlouisfed.org/series/SMU06417400500000001SA
### All Employee

```python
#https://fred.stlouisfed.org/series/SMU06417400500000001SA
names = ['DATE','allEmployee']
df = pd.read_csv('chaodata/allemployee.csv',names=names,skiprows=1)
dates = pd.date_range('1990-01', periods=len(df), freq='M')
df = Series(df.allEmployee.values, index=dates)
sns.set_style("darkgrid")

plt.plot(df)
plt.title('allEmployee, San Diego County')
plt.xlabel('Year')
plt.ylabel('thousands of person')
plt.show()
```

## https://www.bls.gov/regions/west/news-release/2016/occupationalemploymentandwages_sandiego_20160726.htm
### CivilianLaborForce

```python
#https://www.bls.gov/regions/west/news-release/2016/occupationalemploymentandwages_sandiego_20160726.htm
names = ['DATE','Labors']
df = pd.read_csv('chaodata/CivilianLaborForce.csv',names=names,skiprows=1)
dates = pd.date_range('1990-01', periods=len(df), freq='M')
df = Series(df.Labors.values, index=dates)
df.plot()
plt.title('CivilianLaborForce, San Diego County')
plt.xlabel('Year')
plt.ylabel('number of labors')
plt.show()
```

## House Price

```python
names = [
    'ObservationDate',
    'HousePrice'
]
df = pd.read_excel('data/All_Transaction_Prices.xls',names=names,skiprows=10)
y=df.HousePrice.as_matrix()
x = range(1975,2017)
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('House Price for each Year')
plt.show()
```

## HomeownershipRate

```python
names = ['ObservationDate','HomeownershipRate']
df = pd.read_excel('data/Ownership_Rates.xls',names=names,skiprows =10)
y=df.HomeownershipRate.as_matrix()
x = range(2009,2016)
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('HomeownershipRate')
plt.title('HomeownershipRate for year')
plt.show()
```

## Rental Vacancy

## Median Income

```python
names = ['ObservationDate','MedianIncome1','MedianIncome']
df = pd.read_excel('data/Median_Income.xls',names =names,skiprows=19)
plt.plot(df.ObservationDate,df.MedianIncome)
plt.xlabel('Year')
plt.ylabel('Median Income')
plt.title('Median Income for year')
plt.show()
```
